```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

https://www.kaggle.com/datasets/gregorut/videogamesales

```python
data = pd.read_csv('vgsales.csv')
print(data.shape)
data.sample(5)
```

```python
data.describe()
```

```python
data.dropna(inplace=True)
data.describe()
```

```python
data['Year'] = data['Year'].astype(int)
data = data[data['Year']< 2016]
```

```python
import plotly

year_sales = data[['Year', 'Global_Sales']].groupby('Year').sum()
year_sales_jp = data[['Year', 'JP_Sales']].groupby('Year').sum()
fig1 = go.Figure()
fig1.add_trace(go.Scatter(x=year_sales.index, y=year_sales['Global_Sales'], line = dict(color = 'green', width = 2),
                    name = 'Всемирные продажи'))

fig1.add_trace(go.Scatter(x=year_sales_jp.index, y=year_sales_jp['JP_Sales'],
                    line = dict(color = 'blue', width = 2),
                    name = 'Продажи в Японии'))


fig1.update_layout(title='Общее количество продаж видеоигр до 2015 года',
                    xaxis_title='Год',
                    yaxis_title='Продажи (в млн)')
fig1.show()
```

```python
fig = go.Figure()

fig.add_trace(go.Histogram(x=data['Year'],
              xbins=go.histogram.XBins(size=2)))

fig.update_layout(title='Количество игр с продажами более 100 000',
                    xaxis_title='Год',
                    yaxis_title='Количество игр')
fig.show()
```
