```python
import pandas as pd
import spacy

nlp = spacy.load("en_core_web_sm")
```

### Загрузка данных

```python
train = pd.read_csv('train.csv')
```

```python
train.head()
```

```python
from tqdm.auto import tqdm
tqdm.pandas()
```

```python
#Получаем лист из частей речи в тексте
train['new_col'] = train['lyric'].progress_apply(lambda x: [token.tag_ for token in nlp(x)])

#Составляем колонки с колличеством частей речи в тексте
tags = set()
for tag_list in train['new_col']:
    for tag in tag_list:
        if tag and tag not in  ["''", '``', ':', ',', '.']:
            tags.add(tag)
tags = list(tags)
for i in tqdm(tags):
    train[i] = train['new_col'].apply(lambda x: x.count(i))
train
```

```python
train.columns
```

```python
train['sum'] = train.iloc[:, 3:].sum(axis=1)
train
```

```python

#Считаем долю частей речи в тексте

for i in tags:
    train[i] = train[i]/train['sum']

#Нормализуем
for column in tags:
    train[column] = (train[column] - train[column].mean()) / (train[column].abs().max())     
train.head()
```

```python
train.drop(['sum', 'lyric', 'new_col'], axis = 1,inplace=True)
train.sample(5)
```

```python
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import precision_score 
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression



X_train, X_test, y_train, y_test = train_test_split(train.drop(columns=['class']), train['class'], test_size=0.2, random_state=42)
```

```python
classifier = svm.SVC() # init classification model
sclaer = StandardScaler()
X_train_scaled = sclaer.fit_transform(X_train)
X_test_scaled = sclaer.transform(X_test)
classifier.fit(X_train_scaled, y_train)
y_pred = classifier.predict(X_test_scaled)

pres = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted') 
f1 = f1_score(y_test, y_pred, average='weighted')
roc_auc = roc_auc_score(y_test, y_pred) 


print('precision = {0}'.format(pres))
print('recall = {0}'.format(recall))
print('F1  = {0}'.format(f1))
print('ROC AUC = {0}'.format(roc_auc))
```

```python
classifier = RandomForestClassifier()
classifier.fit(X_train_scaled, y_train) # train the model
y_pred = classifier.predict(X_test_scaled) # make predictions for test vectors

pres = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted') 
f1 = f1_score(y_test, y_pred, average='weighted') # calculate f1 score
roc_auc = roc_auc_score(y_test, y_pred) 

print('precision = {0}'.format(pres))
print('recall = {0}'.format(recall))
print('F1  = {0}'.format(f1))
print('ROC AUC = {0}'.format(roc_auc))
```
